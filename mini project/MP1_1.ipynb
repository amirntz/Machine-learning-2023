1_1 Produce a data with 1 sample, 2 classes and 2 features using Sklearn.datasets

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
X,y = make_classification(n_samples= 1000,n_classes=2,n_features=2,n_redundant=0,n_clusters_per_class=1,class_sep=2,random_state=13)
plt.scatter(X[:,0],X[:,1],c=y)
X,y
```

1_2 Separate the two classes in the previous data by using at least two categories of Python's readiness and considering the appropriate transfers.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression,SGDClassifier
X_train,X_test,y_train,y_test = train_test_split(X , y , test_size=0.2,random_state=13)
model_1 = LogisticRegression(solver='sag',max_iter=50,random_state=13)
model_1.fit(X_train , y_train)
model_1.predict(X_test)
a=model_1.score(X_train,y_train)
b=model_1.score(X_test , y_test)
print(f'train_score={a} , test_score={b}')
```

```python
model_2 = SGDClassifier(loss ='log_loss',max_iter=50,random_state = 13)
model_2.fit(X_train , y_train)
model_2.predict(X_test)
a=model_2.score(X_train,y_train)
b=model_2.score(X_test , y_test)
print(f'train_score={a} , test_score={b}')
```

1_3 Show the boundaries and areas of decision -making from your training model along with the samples in a chart. If you can show the samples that are incorrectly classified in different shapes.

```python
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
y_pred = model_1.predict(X)

misclassified_points = X[y != y_pred]


plt.figure(figsize=(10, 5))
plot_decision_regions(X, y, clf=model_1, legend=2)


plt.scatter(misclassified_points[:, 0], misclassified_points[:, 1], s=100, marker='X', c='red', label='Misclassified')


plt.title('Decision Regions with Misclassified Points')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
def sigmoid(x):
  return 1/(1 + np.exp(-x))
```

```python
def logistic_regression(x , w):
  y_hat = sigmoid(x @ w)
  return y_hat
```

```python
def bce(y , y_hat):
  loss = -(np.mean(y*np.log(y_hat) + (1-y)*np.log(1-y_hat)))
  return loss
```

```python
def gradient(x , y , y_hat):
  grads = (x.T @ (y_hat - y)) / len(y)
  return grads
```

```python
def gradient_descent(w,eta,y_hat):
  w -= eta*grads
  return w
```

```python
def accuracy(y , y_hat):
  acc = np.sum(y == np.round(y_hat))/len(y)
  return acc
```
